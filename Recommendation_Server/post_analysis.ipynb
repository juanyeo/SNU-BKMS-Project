```python
import pandas as pd

df = pd.read_csv('stackoverflow_posts_questions_dbms.csv')
print(df.shape)
df.head()
```

```python
def replaced_str(mystring, elimination_list = [], exchange_list = [], inside_list = []):
    
    def exchange_str(mystring, exchange_list):
        for key, val in exchange_list:
            mystring = mystring.replace(key, val)
        return mystring

    def delete_str(mystring, elimination_list):
        for e in elimination_list:
            mystring = mystring.replace(e, '')
        return mystring
            
    def delete_inside(mystring, inside_list):
        
        for start, end in inside_list:
            init_mystring = ''
            for _ in range(5):
                init_mystring = mystring
                s = mystring.find(start)
                e = mystring.find(end) + len(end)
                mystring = mystring[:s] + mystring[e:] 
                if init_mystring != mystring:
                    break
        return mystring

    mystring = delete_str(mystring, elimination_list)
    mystring = exchange_str(mystring, exchange_list)
    mystring = delete_inside(mystring, inside_list)
        
    return mystring    

elimination_list = ['<p>', '</p>']
inside_list = [('<code>', '</code>'), ('<pre>', '</pre>')]

for i, b in enumerate(df['body']):
    b = replaced_str(b, elimination_list=elimination_list, inside_list=inside_list)
    df['body'][i] = b
```
